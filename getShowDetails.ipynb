```python
import re
from datetime import datetime, timedelta

text = """!SCHEDULE:May 6–11: Tuesday @7pm, Wednesday @2pm and 7:30pm, Thursday @7:30pm, Friday @7pm, Saturday 2pm and 8pm, Sunday @3pmMay 20–25: Tuesday @7pm, Wednesday @7pm, Thursday @7pm, Friday @7pm, Saturday 2pm and 8pm, Sunday @2pm and 7:30pmMay 27–June 1: Tuesday @7pm, Wednesday @2pm and 7:30pm, Thursday @7:30pm, Friday @7pm, Saturday 2pm and 8pm, Sunday @3pm."""

# Regex to split date ranges and their corresponding schedules
blocks = re.findall(r"(May|June) (\d{1,2})–(?:(May|June) )?(\d{1,2}): (.*?)(?=(May|June) \d{1,2}–|$)", text)

# Helper to get full date objects for a range
def generate_date_range(start_month, start_day, end_month, end_day):
    months = {"May": 5, "June": 6}
    start_date = datetime(2024, months[start_month], int(start_day))
    end_date = datetime(2024, months[end_month or start_month], int(end_day))
    delta = (end_date - start_date).days
    date_map = {}
    for i in range(delta + 1):
        d = start_date + timedelta(days=i)
        date_map[d.strftime("%A")] = f"{d.strftime('%B')} {int(d.strftime('%d'))}"
    return date_map


# Extract schedule from each block
final_results = []
for block in blocks:
    start_month, start_day, end_month, end_day, schedule = block[:5]
    date_map = generate_date_range(start_month, start_day, end_month, end_day)

    # Find (Day, time1, optional time2)
    matches = re.findall(r"(Tuesday|Wednesday|Thursday|Friday|Saturday|Sunday)[ @]*(\d{1,2}(?::\d{2})?pm)(?: and (\d{1,2}(?::\d{2})?pm))?", schedule)

    for day, time1, time2 in matches:
        date = date_map.get(day)
        if date:
            final_results.append((date, day, time1))
            if time2:
                final_results.append((date, day, time2))

# Print results
for r in final_results:
    print(r)
```

```python
def extract_dates(X):
    text = X["description"]
    blocks = re.findall(r"(May|June) (\d{1,2})–(?:(May|June) )?(\d{1,2}): (.*?)(?=(May|June) \d{1,2}–|$)", text)

    # Helper to get full date objects for a range
    def generate_date_range(start_month, start_day, end_month, end_day):
        months = {"May": 5, "June": 6}
        start_date = datetime(2024, months[start_month], int(start_day))
        end_date = datetime(2024, months[end_month or start_month], int(end_day))
        delta = (end_date - start_date).days
        date_map = {}
        for i in range(delta + 1):
            d = start_date + timedelta(days=i)
            date_map[d.strftime("%A")] = f"{d.strftime('%B')} {int(d.strftime('%d'))}"
        return date_map


    # Extract schedule from each block
    final_results = []
    for block in blocks:
        start_month, start_day, end_month, end_day, schedule = block[:5]
        date_map = generate_date_range(start_month, start_day, end_month, end_day)

        # Find (Day, time1, optional time2)
        matches = re.findall(r"(Tuesday|Wednesday|Thursday|Friday|Saturday|Sunday)[ @]*(\d{1,2}(?::\d{2})?pm)(?: and (\d{1,2}(?::\d{2})?pm))?", schedule)

        for day, time1, time2 in matches:
            date = date_map.get(day)
            if date:
                final_results.append([date, day, time1])
                if time2:
                    final_results.append([date, day, time2])
    return {X["show_id"]: final_results}
```
